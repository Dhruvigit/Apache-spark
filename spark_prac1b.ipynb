```python
from pyspark import SparkContext,SparkConf
```

```python
conf=SparkConf().setAppName("test").setMaster("local")
```

```python
conf
```

```python
sc=SparkContext(conf=conf)
```

```python
sc
```

```python
celsius_temp=[-2,0,20,30,40,100]
```

```python
rdd=sc.parallelize(celsius_temp)
```

```python
fahrenheit_rdd=rdd.map(lambda c:(c,(c*9/5)+32))
```

```python
fahrenheit_temp=fahrenheit_rdd.collect()
```

```python
for c,f in fahrenheit_temp:
    print(f"Celsius:{c},Fahrenheit:{f}")
```

```python
sc.stop()
```
