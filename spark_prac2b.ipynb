```python
from pyspark.sql import SparkSession
```

```python
from pyspark.sql.functions import col
```

```python
#initialize spark session
spark2=SparkSession.builder.appName('WineQualityAnalyzer').getOrCreate()
```

```python
spark2
```

```python
df=spark2.read.format('csv').option('header','true').option('inferSchema','true').load('winequality-red.csv')
```

```python
df.show(5)
```

```python
#filter with a quality rating of 7 or higher
df_filtered=df.filter(col('quality')>=7)
```

```python
df_filtered.show()
```

```python
spark2.stop()
```
